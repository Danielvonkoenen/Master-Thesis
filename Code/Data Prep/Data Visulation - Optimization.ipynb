```python
# Day Energy Consumption with PVs

charge_energy_over_the_day_withPV9 = []


# read all the 100 exported csv files from the simulation about the chargers
# get the data for each hour
for i in range(1,101):
    chargedata9 = pd.read_csv('9 chargeenergywithPV +%i.csv'%(i))
    chargedata9['0'] = pd.to_datetime(chargedata9['0'], unit='s')
    for i in range(0,24):
        charge_energy_over_the_day_withPV9.append([i,sum(chargedata9['1'].loc[(chargedata9['0'].dt.hour) == i])])
```

```python
charge_energy_over_the_day_withPV9 = pd.DataFrame(charge_energy_over_the_day_withPV9)
tmp9 = []


# summarize the data for each hour
for i in range(0,24):
        tmp9.append([i,sum(charge_energy_over_the_day_withPV9[1].loc[(charge_energy_over_the_day_withPV9[0]) == i])])
        
tmp9 = pd.DataFrame(tmp9) 
```

```python
# Calculate average per hour
tmp9[1] = tmp9[1].div(100).round(2)   

# Convert KW to Megawatt
tmp9[1] = tmp9[1].div(1000).round(2) 
```
